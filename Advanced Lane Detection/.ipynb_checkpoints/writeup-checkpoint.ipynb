**Advanced Lane Finding Project**

The goals / steps of this project are the following:

* Compute the camera calibration matrix and distortion coefficients given a set of chessboard images.
* Apply a distortion correction to raw images.
* Use color transforms, gradients, etc., to create a thresholded binary image.
* Apply a perspective transform to rectify binary image ("birds-eye view").
* Detect lane pixels and fit to find the lane boundary.
* Determine the curvature of the lane and vehicle position with respect to center.
* Warp the detected lane boundaries back onto the original image.
* Output visual display of the lane boundaries and numerical estimation of lane curvature and vehicle position.

### Camera Calibration

#### 1. Briefly state how you computed the camera matrix and distortion coefficients. Provide an example of a distortion corrected calibration image.

The code for this step is in the cell of 'Step 2. Camera Calibration' of IPython notebook 'AdvancedLaneDetection.ipynb'.

I start by preparing "object points", which will be the (x, y, z) coordinates of the chessboard corners in the world. Here I am assuming the chessboard is fixed on the (x, y) plane at z=0, such that the object points are the same for each calibration image.  Thus, `objp` is just a replicated array of coordinates, and `objpoints` will be appended with a copy of it every time I successfully detect all chessboard corners in a test image.  `imgpoints` will be appended with the (x, y) pixel position of each of the corners in the image plane with each successful chessboard detection.  

The sample chessboard images provided have 6x9 numbers of corners. The following image shows the corner points on chessboard images. There are three images we can not detect corners because they are too distorted and zoomed in too much that we can only view less than 6x9 corners.
![alt text][image1]


I then used the output `objpoints` and `imgpoints` to compute the camera calibration and distortion coefficients using the `cv2.calibrateCamera()` function. I applied this distortion correction to the test image (calibration image 1) using the `cv2.undistort()` function and obtained this result: 
![alt text][image2]

### Discussion

#### 1. Briefly discuss any problems / issues you faced in your implementation of this project.  Where will your pipeline likely fail?  What could you do to make it more robust?
From 38 to 44 seconds, when the road changes from black aspalt to brighter concrete, the performance of the script droped. Because the lane detection gets poors. To compensate that, I create a Line class to store the previous calculated poly-fit coefficients by using a queue, i.e 'current_fit'. I set the maximum length of this queue to be 5, so the calculation will always look maximum 5 previous frames to decide if the new fit coefficient should be kept or thrown away, by using a threshold value for the difference of coefficents.

Then I decide to set the thresholds by comparing the fit coefficient of multiple frames, in order to get the magnitude for y0, y1 and y2. For example, the following fit coefficients of two frames gives us the difference between y0, y1 and y2 at corrspondingly: 

Difference at left lane y0,y1,y2:  [  2.33765514e-04   2.63533239e-01   7.07220789e+01]

Difference at right lane y0,y1,y2:  [  4.00178000e-05   4.40609398e-02   7.98937195e+00]

I ran a couple of simulation with printouts of the difference, providing them with a safe margin, I decide to choose 0.001, 1 and 100.

Although the pipeline is good enough for the project video, it is far from robust to deal with real world situations. To improve the quality of challenge videos, I have to go back to adjust the color space thresholdFor example, how to deal with really curvy roads, unmarked lane or poorly marked lane, or other complicated situation such as vehicle cutting into the lane or street road scenarios.
